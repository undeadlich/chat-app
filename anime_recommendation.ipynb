```python
import numpy as np
import pandas as pd
```

```python
anime = pd.read_csv("anime-dataset-2023.csv")
```

```python
anime.head()
```

```python
anime = anime[['anime_id','English name','Genres','Synopsis','Studios']]
anime['Genres']
```

```python
(anime == 'UNKNOWN').sum()
anime = anime.replace('UNKNOWN', pd.NA)
(anime['Synopsis'] == 'No description available for this anime.').sum()
anime = anime.replace('No description available for this anime.', pd.NA)
anime.isna().sum()
anime.dropna(inplace=True)
```

```python
anime = anime.reset_index(drop=True)
anime
```

```python
def convert(obj):
    L = []
    L.append(obj)
    return L
```

```python
anime['Studios'] = anime['Studios'].apply(convert)
anime
```

```python
anime['Genres'] = anime['Genres'].apply(lambda x:x.split(','))
anime['Synopsis'] = anime['Synopsis'].apply(lambda x:x.split(' '))

```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
anime['Genres'] = anime['Genres'].apply(collapse)
anime['Studios'] = anime['Studios'].apply(collapse)
anime['Synopsis'] = anime['Synopsis'].apply(collapse)
anime
```

```python
anime['tags'] = anime['Genres'] + anime['Studios'] + anime['Synopsis']
new_anime = anime.drop(columns=['Genres','Studios','Synopsis'])
```

```python
new_anime['tags'] = new_anime['tags'].apply(lambda x: " ".join(x))
new_anime['tags']

```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_anime['tags'] = new_anime['tags'].apply(stem)
```

```python
new_anime
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vector = cv.fit_transform(new_anime['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
#new_anime[new_anime['English name']== 'Cowboy Bebop']
```

```python
def recommend(anime):
  ind =  new_anime[new_anime['English name'] == anime].index[0]
  print(ind)
  distances = sorted(list(enumerate(similarity[ind])),reverse=True,key = lambda x: x[1])
  for i in distances[1:6]:
    print(new_anime.iloc[i[0]]['English name'])
```

```python
recommend('Mushoku Tensei: Jobless Reincarnation')
```
